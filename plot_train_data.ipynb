```python
train_labels_array = np.loadtxt('train_labels.csv', delimiter=',', dtype=str)
```

```python
train_labels_array.shape
```

```python
train_labels_array[:3,:]
```

```python
train_labels_dict = {train_labels_array[i,0]: train_labels_array[i,1] for i in range(train_labels_array.shape[0])}
```

```python
# train_labels_dict
```

```python
plt.figure(figsize=(6,9))
plt.suptitle("target: {}".format(train_labels_dict["0024012d1431fbc"]))

for i in range(train_0024012d1431fbc.shape[0]):
    plt.subplot(3,2,i+1)
    plt.imshow(train_0024012d1431fbc[i])
    # plt.ylabel("Time")
    # plt.xlabel("Relative Frequency")
    plt.title("cadence {}".format(i))
plt.tight_layout()
# plt.show()
```
